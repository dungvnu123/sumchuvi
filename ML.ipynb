## Train , valatidition, test data_base

```python
train, valid , test = np.split(data_df.sample(frac=1), [int(0.6*len(data_df)), int(0.8*len(data_df))])
```

```python

def mising_value(data_df, check=False):
    # Lấy X: tất cả các cột trừ cột cuối (classs)
    x = data_df[data_df.columns[:-1]].values

    # ❌ Lỗi gốc ở đây: bạn dùng data_df.columns['classs'] là sai
    # ✅ Dùng đúng cách: gọi trực tiếp theo tên cột
    y = data_df['classs'].values

    # Chuẩn hóa
    scaler = StandardScaler()
    x = scaler.fit_transform(x)

    # Resample nếu cần
    if check:
        ros = RandomOverSampler()
        x, y = ros.fit_resample(x, y)

    # Ghép lại thành 1 mảng (x + y làm cột cuối)
    data = np.hstack((x, np.reshape(y, (-1, 1))))

    return data, x, y
```

```python
train, x_train, y_train = mising_value(train, check = True)
valid, x_valid, y_valid = mising_value(valid, check = False)
test, x_test, y_test = mising_value(test, check = False)
```

## Neuron network


```python
import tensorflow as tf
```

```python
def plot_history(history):
    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(10, 4))

    # Loss
    ax1.plot(history.history['loss'], label='train')
    if 'val_loss' in history.history:
        ax1.plot(history.history['val_loss'], label='val')
    ax1.set_xlabel('Epoch')
    ax1.set_ylabel('Loss')
    ax1.grid(True)
    ax1.legend()

    # Accuracy
    if 'accuracy' in history.history:
        ax2.plot(history.history['accuracy'], label='train')
    if 'val_accuracy' in history.history:
        ax2.plot(history.history['val_accuracy'], label='val')
    ax2.set_xlabel('Epoch')
    ax2.set_ylabel('Accuracy')
    ax2.grid(True)
    ax2.legend()

    plt.tight_layout()
    plt.show()
```

```python
def train_model(x_train, y_train, num_nodes , dropout_pd, lir, batchs_size, eposhs):
  nn_model = tf.keras.Sequential([
      tf.keras.layers.Dense(num_nodes, activation='relu', input_shape=(10,)),
      tf.keras.layers.Dropout(dropout_pd),
      tf.keras.layers.Dense(num_nodes, activation='relu'),
      tf.keras.layers.Dropout(dropout_pd),
      tf.keras.layers.Dense(1, activation='sigmoid')
  ])
  nn_model.compile(optimizer=tf.keras.optimizers.Adam(learning_rate=lir),
                   loss='binary_crossentropy',
                   metrics=['accuracy'])
  history = nn_model.fit(
      x_train, y_train,
      epochs=eposhs,
      batch_size=batchs_size,
      validation_split=0.2,
      verbose=1

  )
  return nn_model, history
```

```python
eposhs = 100
for num_node in [16, 32, 64]:
  for dropout_pd in [0, 0.2]:
    for lir in [0.001, 0.002, 0.0005]:
      for batchs_size in [32, 64, 128]:
        print(f"Training with nodes={num_node}, dropout={dropout_pd}, lr={lir}, batch={batchs_size}")
        nn_model, history = train_model(x_train, y_train, num_node, dropout_pd, lir, batchs_size, eposhs)
        plot_history(history)
```
